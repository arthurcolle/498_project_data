```python
# Render our plots inline
%matplotlib inline

# Import several libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Import linear regression and the diabetes dataset
from sklearn.linear_model import LinearRegression
```

```python
nge_24 = pd.read_csv('nge_24.csv')
nge_12 = pd.read_csv('nge_12.csv')
nge_diff = pd.read_csv('nge_diff.csv')

size12 = len(nge_12['score'])
size24 = len(nge_24['score'])
sizeDiff = len(nge_diff['score'])

size12_20 = int(size12*.2)
size24_20 = int(size24*.2)
sizeDiff_20 = int(sizeDiff*.2)

nge_12x_train = nge_12['score'][:-size12_20]
nge_12x_train = nge_12x_train.reshape((nge_12x_train.shape[0], -1))
nge_12x_test = nge_12['score'][-size12_20:]
nge_12x_test = nge_12x_test.reshape((nge_12x_test.shape[0], -1))
nge_12y_train = nge_12['percent_change'][:-size12_20]
nge_12y_train = nge_12y_train.reshape((nge_12y_train.shape[0], -1))
nge_12y_test = nge_12['percent_change'][-200:]

nge_24x_train = nge_24['score'][:-size24_20]
nge_24x_train = nge_24x_train.reshape((nge_24x_train.shape[0], -1))
nge_24x_test = nge_24['score'][-size24_20:]
nge_24x_test = nge_24x_test.reshape((nge_24x_test.shape[0], -1))
nge_24y_train = nge_24['percent_change'][:-size24_20]
nge_24y_train = nge_24y_train.reshape((nge_24y_train.shape[0], -1))
nge_24y_test = nge_24['percent_change'][-200:]


nge_diffx_train = nge_diff['score'][:-sizeDiff_20]
nge_diffx_train = nge_diffx_train.reshape((nge_diffx_train.shape[0], -1))
nge_diffx_test = nge_diff['score'][-sizeDiff_20:]
nge_diffx_test = nge_diffx_test.reshape((nge_diffx_test.shape[0], -1))
nge_diffy_train = nge_diff['percent_change'][:-sizeDiff_20]
nge_diffy_train = nge_diffy_train.reshape((nge_diffy_train.shape[0], -1))
nge_diffy_test = nge_diff['percent_change'][-200:]

# Create linear regression object
regr1 = LinearRegression()
regr2 = LinearRegression()
regr3 = LinearRegression()

# Train the model using the training sets
regr1.fit(nge_12x_train, nge_12y_train)
regr2.fit(nge_24x_train, nge_24y_train)
regr3.fit(nge_diffx_train, nge_diffy_train)


predicted_targets12 = regr1.predict(nge_12x_test)
predicted_targets24 = regr2.predict(nge_24x_test)
predicted_targetsDiff = regr3.predict(nge_diffx_test)
```
