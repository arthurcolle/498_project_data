```python
# Render our plots inline
%matplotlib inline

# Import several libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Import linear regression and the diabetes dataset
from sklearn.linear_model import LinearRegression
```

```python
size = len(nge_24['score'])
size20 = int(size*.2)


nge_24x_train = nge_24['score'][:-size20]
nge_24x_train = nge_24x_train.reshape((nge_24x_train.shape[0], -1))
nge_24x_test = nge_24['score'][-size20:]
nge_24x_test = nge_24x_test.reshape((nge_24x_test.shape[0], -1))

nge_24y_train = nge_24['percent_change'][:-size20]
nge_24y_train = nge_24y_train.reshape((nge_24y_train.shape[0], -1))
nge_24y_test = nge_24['percent_change'][-20:]

# Create linear regression object
regr = LinearRegression()

# Train the model using the training sets
regr.fit(nge_24x_train, nge_24y_train)

predicted_targets = regr.predict(nge_24x_test)
```

```python
nge_24 = pd.read_csv('nge_24.csv')
nge_12 = pd.read_csv('nge_12.csv')
nge_diff = pd.read_csv('nge_diff.csv')

fig = plt.figure(figsize=(10, 8))
ax1 = fig.add_subplot(111)
ax1.scatter(nge_24['score'], nge_24['percent_change'], c='blue', label='24hr')
ax1.scatter(nge_12['score'], nge_12['percent_change'], c='green', label='12hr')
ax1.scatter(nge_diff['score'], nge_diff['percent_change'], c='red', label='diff')
ax1.plot(nge_24x_test, predicted_targets, c='black', label='regr', linewidth=5.0, pickradius=10.0)

plt.legend(loc='upper left');
plt.xlabel("Sentiment", fontsize=20)
plt.ylabel("Percent Change", fontsize=20)
plt.title("NGE", fontsize=30)
plt.show()
```
